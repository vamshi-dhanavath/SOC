*  Bivariate Analysis:

  *  Create scatter plots, correlation matrices, and pair plots to explore relationships between numerical variables.



```python
plt.scatter(dataframe['Sales'],dataframe['Profit'])
plt.title('profit vs sales')
plt.xlabel('sales')
plt.ylabel('profit')
plt.show()
```

```python
plt.scatter(dataframe['Discount'],dataframe['Profit'])
plt.title('profit vs discount')
plt.xlabel('discount')
plt.ylabel('profit')
plt.show()
```

```python
#correlation_matrix
corr_matrix=dataframe[['Sales','Profit','Quantity','Discount']].corr()
```

```python
import seaborn as sns

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix for Sales,Profit,Quantity,Discount')
plt.show()
```

*    Use bar plots and box plots to explore relationships between numerical and categorical variables (e.g., Sales by Category, Profit by Region).

```python
sns.boxplot(x='Category',y='Sales',data=dataframe)
plt.title('box plot of sales by category')
plt.show()
```

```python
sns.boxplot(x='Region',y='Profit',data=dataframe)
plt.xticks(rotation=90)
plt.title('box plot of profit by region')
plt.show()
```

## Feature Engineering

```python
dataframe['Profit Margin'] = dataframe['Profit'] / dataframe['Sales'] * 100                        #profit margin
dataframe['Sales Category'] = pd.cut(dataframe['Sales'], bins=[0, 100, 200, float('inf')], labels=['Low', 'Medium', 'High'])         #sales category

plt.figure(figsize=(10, 6))
sns.barplot(x='Sales Category', y='Sales', data=dataframe)
plt.title('sales by Sales Category')
plt.show()
```

```python
#order processig  time
dataframe['Order Processing Time'] = (dataframe['Ship Date'] - dataframe['Order Date']).dt.days

sns.barplot(x='Order Processing Time',y='Sales',data=dataframe)
plt.title('Order Processing Time')
plt.show()
```
